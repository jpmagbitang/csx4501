## Introduction to the Dataset

This dataset is gathered from the website: https://www.kaggle.com/datasets/robikscube/pga-tour-golf-data-20152022. <br> Within this dataset is a is golfer data from players on the PGA Tour from 2021. Some variables involved include each players results at different tournaments, strokes gained from different levels, whether they were cut or played all four rounds, etc. It also has stats from different sportsbooks websites such as Fanduel and Draft Kings. However, the made focus for the data is how well each player had performed in each tournament. 

## Import Modules for Pandas and Matplotlib 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Import the data

The data is found in the CSV file, "PGAdata2021.csv"

### Plot of Total Strokes

```python
df['strokes'].plot()
```

### Plot of Par of Hole 

```python
df['hole_par'].plot()
```

### Plot of Hole_Par vs Strokes

```python
df.plot(x='hole_par', y='strokes', style='.')
```

### Plot of the top 10 Players with the Most 2022 Season Wins

```python
plt.figure(figsize=(15,8))
df.loc[df['pos'] == 1]['player'].value_counts().head(10).plot(kind = 'bar')
plt.show()
```

### Plot of the top 10 Players with 2nd Place Finishes

```python
plt.figure(figsize=(15,8))
df.loc[df['pos'] == 2]['player'].value_counts().head(10).plot(kind = 'bar')
plt.show()
```
