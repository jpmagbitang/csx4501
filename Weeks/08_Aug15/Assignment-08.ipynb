# NLP Visualizations of Macbeth and Romeo and Juliet

Execute the following cell for important imports.

```python
import requests
import matplotlib.pyplot as plt
import nltk
from nltk.tokenize import word_tokenize
from nltk.sentiment import vader
from nltk.corpus import stopwords
from nltk.corpus import opinion_lexicon
from nltk.stem.porter import PorterStemmer

#will be used for the barplots later on
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

Execute the following cell to get the text of "Macbeth" and "Romeo and Juliet" from [Project Gutenberg](https://www.gutenberg.org/) 

```python
target_url = "https://www.gutenberg.org/files/1533/1533-0.txt"
response = requests.get(target_url)
macbeth_string = response.text
macbeth_string = macbeth_string.split('*** START OF THE PROJECT GUTENBERG EBOOK MACBETH ***')[1]
macbeth_string = macbeth_string.split('*** END OF THE PROJECT GUTENBERG EBOOK MACBETH ***')[0]
macbeth_string = macbeth_string.lower()

target_url = "https://www.gutenberg.org/files/1513/1513-0.txt"
response = requests.get(target_url)
r_j_string = response.text
r_j_string = r_j_string.split('*** START OF THE PROJECT GUTENBERG EBOOK ROMEO AND JULIET ***')[1]
r_j_string = r_j_string.split('*** END OF THE PROJECT GUTENBERG EBOOK ROMEO AND JULIET ***')[0]
r_j_scenes = [i.lower() for i in r_j_string.split('SCENE')[1:]]
```

You now have the following variables:
* `macbeth_string`: this string contains the full text of "Macbeth"
* `r_j_string`: this string contains the full text of "Romeo and Juliet"
* `r_j_scenes`: the text from each scene of "Romeo and Juliet" is included in this Python list, with one scene per list item

Make two new variables:
* a Python list of stopwords
* a variable for using VADER's SentimentIntensityAnalyzer

```python
myStopwords = stopwords.words('english')
sia = vader.SentimentIntensityAnalyzer()
```

## Macbeth

Make a list of the words in Macbeth by tokenizing the macbeth_string:

```python
words = word_tokenize(macbeth_string)
```

Next, make a list of positive words and a list of negative words using the `polarity_scores` method of VADER's SentimentIntensityAnalyzer.

Write a for loop that iterates over your Macbeth word list, and:
* if the compound score of a word is positive, add the word to the positive-word list
* if the compound score of a word is negative, add the word to the negative-word list

```python
wordsposi = []
wordsnegi = []
wordsna = []
# the na variable deals with punctionations that would be included if not specified

for i in words:
    if sia.polarity_scores(i)['compound'] > 0:
        wordsposi.append(i)
    elif sia.polarity_scores(i)['compound'] < 0:
        wordsnegi.append(i)
    else:
        wordsna.append(i)

```

For any list `mylist`, you can get the word frequencies of words in `mylist` using `nltk.FreqDist(mylist)`, and you can get the `n` most common words with `nltk.FreqDist(mylist).most_common(n)`.

Use these commands to:
1. find the 15 most common negative words in Macbeth and the 15 most common positive words
2. make a horizontal bar plot similar to the plots in the following
3. comment on similarities and differences to the Macbeth plot below

(Plot taken from https://peerchristensen.netlify.app/post/fair-is-foul-and-foul-is-fair-a-tidytext-entiment-analysis-of-shakespeare-s-tragedies/)

![Pos-Neg-Words in Shakespeare](https://peerchristensen.netlify.com/post/2018-06-07-fair-is-foul-and-foul-is-fair-a-tidytext-entiment-analysis-of-shakespeare-s-tragedies_files/figure-html/unnamed-chunk-19-1.png)

```python
# Code:
print(nltk.FreqDist(wordsposi).most_common(15))
print(nltk.FreqDist(wordsnegi).most_common(15))
```

```python
# turning most common into its own variables 
mostcommonp = nltk.FreqDist(wordsposi).most_common(15)
mostcommonn = nltk.FreqDist(wordsnegi).most_common(15)

# adding variables word and ammount into the dataframe
dfposi = pd.DataFrame(mostcommonp, columns = ['word','amount'])
dfnegi = pd.DataFrame(mostcommonn, columns = ['word','amount'])

# constructing horizontal barplots
plt.subplots(figsize=(16, 6))

plt.subplot(1, 2, 1)
sns.barplot(x='amount', y='word', data = dfposi, color='orange')
plt.xlabel('Amount',fontsize=16);
plt.ylabel('Word', fontsize=16);
plt.title('Common 15 Positive', fontsize=16, weight = 'bold')

plt.subplot(1, 2, 2)
sns.barplot(x='amount', y='word', data = dfnegi, color = 'purple')
plt.xlabel('Amount',fontsize=16);
plt.ylabel('')
plt.title('Common 15 Negative', fontsize=16,weight='bold')
plt.show()
```

**Looking at the two graphs, for the positive side, good still has the most counts however, when looking at the negative graph, the words 'no' and 'witch' pass up fear to have the most counts for the negative words.**

## Romeo and Juliet

We are now going to plot the evolution of sentiment in Romeo and Juliet.

This is motivated by the following plot (also from the afore-mentioned link):
![Sentiment Evolution in Shakespeare](https://peerchristensen.netlify.com/post/2018-06-07-fair-is-foul-and-foul-is-fair-a-tidytext-entiment-analysis-of-shakespeare-s-tragedies_files/figure-html/unnamed-chunk-11-1.png)

We are going to make a couple different plots.

First:
* Write a for loop that iterate over the scenes in Romeo and Juliet (`r_j_scenes`), and for each scene:
  * Make a list of the scene's words by tokenizing the string
  * Calculate the total scene sentiment score by summing the compound scores of all the words in that scene
* Make a bar plot that has the scene number on the horizontal axis and the sentiment scores as the bar heights.
* Comment on its resemblance to the Romeo and Juliet plot above.

```python
#for loop that gets the Sentiment score for the the given scene
def SentimentScoreSum(scene):
    words = [w for w in word_tokenize(scene)]
    sentimentscore = 0
    for word in words:
        sentimentscore = sentimentscore + sia.polarity_scores(word)['compound']
    return sentimentscore

# loop that iterates over the Romeo and Juliet Scene:
RJscore = []
for scene in r_j_scenes:
    RJscore.append(SentimentScoreSum(scene))

# creating a dataframe for Romeo and Juliet Scenes and each scene's score
scenenumber = []
for i in range(len(r_j_scenes)):
    scenenumber.append('Scene' + str(i+1))
    
# combines the RJscore variable and the scenenumber variable into one list then creates a dataframe    
combinedlist = list(zip(scenenumber, RJscore))
dfSentiSum = pd.DataFrame(combinedlist, columns=['Scene', 'Score'])


#bar plot that shows the sentiment scores for this grouping
plt.subplots(figsize=(16, 6))
sns.barplot(x='Scene', y='Score', data = dfSentiSum)
plt.xlabel('Scene',fontsize=16)
plt.ylabel('Sentiment Score', fontsize=16)
plt.title('Romeo and Juliet Sentiment Scores', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

**In the Romeo and Juliet plot above, the scentiment scores are back and forth starting from the negative side and it seems as if the scores are greater. With the new plot, up until scene 13 the scores are positive and back and forth after. The new plot also has less scores for each scene, while the original plot had more depths and shading.**



Second:
* Write a for loop that iterate over the scenes in Romeo and Juliet (`r_j_scenes`), and for each scene:
  * Make a list of the scene's words by tokenizing the string
  * Calculate the total scene sentiment score by **adding 1 to the score for each positive word (compound_score > 0) and substracting 1 from the score for each negative word (compound_score < 0)**
    * This total is the difference between the positive and negative word counts
* Make a bar plot that has the scene number on the horizontal axis and the sentiment scores as the bar heights.
* Does this change the plot relative to the previous method of calculating the total scene sentiment score?

```python
# for loop for to find that sentiment score for the positive and negative words in the scene
def SentimentScore2(scene):
    words = [w for w in word_tokenize(scene)]
    sentimentscore = 0
    for word in words:
        if sia.polarity_scores(word)['compound']> 0:
            sentimentscore = sentimentscore + 1
        elif sia.polarity_scores(word)['compound'] < 0:
            sentimentscore = sentimentscore - 1
    return sentimentscore

# loop that iterates over the Romeo and Juliet Scene
RJscore = []
for scene in r_j_scenes:
    RJscore.append(SentimentScore2(scene))

# creating a dataframe for Romeo and Juliet Scenes and each scene's score
scenenumber = []
for i in range(len(r_j_scenes)):
    scenenumber.append('Scene' + str(i+1))

# combines the RJscore variable and the scenenumber variable into one list then creates a dataframe
combinedlist = list(zip(scenenumber, RJscore))
dfSenti2 = pd.DataFrame(combinedlist, columns=['Scene', 'Score'])


#bar plot that shows the sentiment scores for this grouping
plt.subplots(figsize=(16, 6))
sns.barplot(x='Scene', y='Score', data = dfSenti2)
plt.xlabel('Scene',fontsize=16)
plt.ylabel('Sentiment Score', fontsize=16)
plt.title('Romeo and Juliet Sentiment Scores', fontsize=18)
plt.xticks(rotation=90)
plt.show()

```

**From the first sentiment score plot created, it had begun on the positive side up until scene 13, now it is more similar to the original plot where there is a negative score in the beginning. The absolute value for the scores for each scene also seem to be greater than the first plot created.**

Third plot:

* Write a for loop that iterate over the scenes in Romeo and Juliet (`r_j_scenes`), and for each scene:
  * Make a list of the scene's words by tokenizing the string
  * Calculate the total scene sentiment score by **adding 1 for each positive word and substracting 1 for each negative word, where the positive/negative word is assessed by its presence in the positive_words and negative_words sets.**
    * This total is the difference between the positive and negative word counts
* Make a bar plot that has the scene number on the horizontal axis and the sentiment scores as the bar heights.
* Does this change the plot relative to the previous plots?

```python
# Loop that calculates the sentiment score using the positive and negative word variables
def SentimentScore3(scene):
    words = [w for w in word_tokenize(scene)]
    sentimentscore = 0
    for word in words:
        if word in positive_words:
            sentimentscore = sentimentscore + 1
        elif word in negative_words:
            sentimentscore = sentimentscore - 1
    return sentimentscore

# loop that iterates over the Romeo and Juliet Scene
RJscore = []
for scene in r_j_scenes:
    RJscore.append(SentimentScore3(scene))

# creating a dataframe for Romeo and Juliet Scenes and each scene's score
scenenumber = []
for i in range(len(r_j_scenes)):
    scenenumber.append('Scene' + str(i+1))

# combines the RJscore variable and the scenenumber variable into one list then creates a dataframe
combinedlist = list(zip(scenenumber, RJscore))
dfSenti3 = pd.DataFrame(combinedlist, columns=['Scene', 'Score'])


#bar plot that shows the sentiment scores for this grouping
plt.subplots(figsize=(16, 6))
sns.barplot(x='Scene', y='Score', data = dfSenti3)
plt.xlabel('Scene',fontsize=16)
plt.ylabel('Sentiment Score', fontsize=16)
plt.title('Romeo and Juliet Sentiment Scores', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

**This new plot is alot more different than the first two plots created. There are now more negative scores and the scores go back and forth from positive to negative straight from the beginning. This set looks as if the split between positive and negative scores is much more even.**



Fourth plot:

* Take the entire string of Romeo and Juliet (`r_j_string`) and tokenize it into words.
* Calculate a total sentiment score for each consecutive group of 500 words.
  * Do this as immediately above by adding 1 for each positive word and substracting 1 for each negative word, where the positive/negative word is assessed by its presence in the positive_words and negative_words sets.
* Make a bar plot that has the group number (0,1,2,3,...) on the horizontal axis and the sentiment scores as the bar heights.
* Does this change the plot relative to the previous plots?

```python
RJtokenized = word_tokenize(r_j_string)
```

```python
# trying to see how many words RJtokenized consists of
len(RJtokenized)
```

```python
# this will tell how many entries the x axis will have
len(RJtokenized)/500
```

```python
# creating the list of entries from groping every 500 words
number_groups = 500
grouplist = [RJtokenized[i:i+number_groups] for i in range(0,32438,500)]

# to confirm the number of entries matches the len(RJtokenized)/500, should be rounded to 65 to include the remainders
len(grouplist)
```

```python
# Loop that calculates the sentiment score for this new grouping
def SentimentScore4(group):
    sentimentscore = 0
    for word in group:
        if word in positive_words:
            sentimentscore = sentimentscore + 1
        elif word in negative_words:
            sentimentscore = sentimentscore - 1
    return sentimentscore

# loop that iterates over the group entries
groupscore = []
for group in grouplist:
    groupscore.append(SentimentScore4(group))

# creating a dataframe for group entries and each group's score
groupnumber = []
for i in range(len(groupscore)):
    groupnumber.append('Group' + str(i+1))

# combines the RJscore variable and the scenenumber variable into one list then creates a dataframe
combinedlist = list(zip(groupnumber, groupscore))
dfSenti4 = pd.DataFrame(combinedlist, columns=['Group', 'Score'])


#bar plot that shows the sentiment scores for this grouping
plt.subplots(figsize=(16, 6))
sns.barplot(x='Group', y='Score', data = dfSenti4)
plt.xlabel('Group Number',fontsize=16)
plt.ylabel('Sentiment Score', fontsize=16)
plt.title('Romeo and Juliet Sentiment Scores', fontsize=18)
plt.xticks(rotation=90)
plt.show()
```

**With this grouping, instead of showing the normal 25 scenes in Romeo and Juliet, it had divided the script of words into groups of 500. By doing this there are now 65 groups and it also seems as there are more negative scores by grouping it this way rather than the previous plots.**
